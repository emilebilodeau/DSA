### 217. Contains Duplicate (easy)

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

```python
# attempt
class Solution:
    def containsDuplicate(self, nums: list) -> bool:

        convert = set(nums)
        
        if len(convert) == len(nums):
            return False
        return True
        
# status = success
```

```python
# solution
class Solution:
    def containsDuplicate(self, nums: list) -> bool:
        hashset = set()

        for n in nums:
            if n in hashset:
                return True
            hashset.add(n)
        return False
```

### 1. Two Sum (easy)

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

```python
# attempt
class Solution:
    def twoSum(self, nums: list, target: int) -> list:
        
        hashmap = {}

        for index, value in enumerate(nums):
            if value in hashmap.keys():
                return [index, hashmap[value]]
            else:
                hashmap[target - value] = index

# status = success
# NOTE: although this answer is very good, it was a bit cheating. I had seen it before and simply had to remember the concept
```

```python
# solution
# big O for this solution
# time complexity: O(n) 
# space complexity: O(n)
class Solution:
    def twoSum(self, nums: list, target: int) -> list:
        prevMap = {}  # val -> index

        for i, n in enumerate(nums):
            diff = target - n
            if diff in prevMap:
                return [prevMap[diff], i]
            prevMap[n] = i
# NOTE: better space complexity for neetcode solution, with still very good time complexity
```

### 141. Linked list Cycle (easy)

Given head, the head of a linked list, determine if the linked list has a cycle in it.

There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. Internally, pos is used to denote the index of the node that tail's next pointer is connected to. Note that pos is not passed as a parameter.

Return true if there is a cycle in the linked list. Otherwise, return false.

```python
# attempt
# big O for this attempt
# time complexity: O(n)
# space complexity: O(n)

# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def hasCycle(self, head: [ListNode]) -> bool:
        
        mySet = set()

        while head:
            if head in mySet:
                return True
            mySet.add(head)
            head = head.next

        return False
    
# status = success
# NOTE: good time complexity
# NOTE: finished this problem in ~7 mins
```

```python
# solution

# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def hasCycle(self, head: ListNode) -> bool:
        slow, fast = head, head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False
```

#### explanation

my own: this problem is a much easier version of the problem 143. Reorder List (basically only the first part of that problem), and the solution here makes use again of the "turtoise and hare" algorithm.

the code here is extremely simple, what's interesting is why this works. begin by creating a slow and fast pointer, both starting at head. then start a while loop that runs while fast and fast.next are not null. if either of those become null that means there is no cycle and you can return false.

then simply increment your slow pointer by 1 in the linked list using slow.next and increment your fast pointer by 2 in the linked list by using fast.next.next. if the slow and fast pointer are ever equal to each other, that means there is a cycle and you can return true.

why this works exactly can easily be explained by the diagram below. essentially, if there is a cycle, each iteration the gap between the slow and fast pointer decreases by 1, and so the maximum number of iteration that the algorithm would need to go through is n if we assume that n the is the entire length of the linked list (this is visualized below by 10 being the gap, +1 being the gap increased by the slow pointer moving up by 1, and -2 being the gap decreased by the fast pointer catching up to the slow pointer). because the gap is decreased by 1 each iteration, that means you are guaranteed that the fast pointer will land exactly where the slow pointer is at some point without skipping over it. a more detailed explanation can be seen on the neetcode video for leetcode problem 141.
